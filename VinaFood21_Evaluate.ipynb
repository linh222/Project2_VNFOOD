```python
from keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
import pickle
import keras
from sklearn.metrics import classification_report,confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

# Load test set

```python
#Load dataset from directory
def Load_Dataset(path,BATCH_SIZE = 64,shuffle = False,Bool = True):
  if Bool:
    datagen = ImageDataGenerator(rescale = 1./225)
  else:
    datagen = ImageDataGenerator()
  X = datagen.flow_from_directory(path,
                                  target_size = (224,224),
                                  batch_size = BATCH_SIZE,
                                  class_mode='categorical',
                                  shuffle = shuffle)
  return X
```

```python
with open('/content/drive/MyDrive/HK5/Project_1/Đồ án số 1 khoa học dữ liệu/Datasets/Dataset Nhóm Mình/classes.pkl', 'rb') as f:
    class_idx = pickle.load(f)
idx_class = dict((v, k) for k, v in class_idx.items())
class_name = class_idx.keys()
```

# Build evaluate function

```python
#confusion matrix
def Confusion_Save(label_true,label_predict,class_name,type_name,model_name):
  cm = confusion_matrix(label_true, label_predict)
  plt.figure(figsize=(12, 12))
  sb.heatmap(cm, annot = True, fmt = 'g', cmap = 'Blues', xticklabels = class_name, yticklabels = class_name)
  plt.xlabel('Predicted label')
  plt.ylabel('True label')
  plt.title('Confusion matrix')
  plt.savefig(f"./Confusion_Matrix/{model_name}/VNFood15/{type_name}_{model_name}")
  plt.show()

def Predict(data,idx_class,class_name,type_name,model,model_name,set_name,header = False):
  predictions = model.predict_generator(data)
  evaluate = model.evaluate_generator(data)

  label_predict = np.argmax(predictions, axis=-1)
  label_true = data.labels
  label_predict_name = [idx_class[elem] for elem in label_predict]
  label_true_name = [idx_class[elem] for elem in label_true]

  report = classification_report(label_true_name, label_predict_name,output_dict=True)
  Confusion_Save(label_true,label_predict,class_name,type_name,model_name)

  df = pd.DataFrame({"Method": [type_name],
                     "Model": [model_name],
                     "Accuracy": [report['accuracy']],
                     "Top_3": [evaluate[2]],
                     "F1-Score": [report['macro avg']['f1-score']]})
  df.to_csv(f"./{set_name}_{type_name}_Evaluate.csv",index = None,header = header,mode = "a")
  print(df)
```

```python
def Evaluate(type_name,model_name,sets,set_name,header = False,idx_class = idx_class,class_name = class_name):
  model = keras.models.load_model(f"./Model/{model_name}/VinaFood21/{type_name}_{model_name}.h5")
  Predict(sets,idx_class,class_name,type_name,model,model_name,header = header)
```
